```python
# Importando bibliotecas
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('./dados/co2.csv')
base.head()
```

```python
# Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type
srn.scatterplot(x = base.conc, y = base.uptake, hue = base.Type)
```

```python
# Seleção de registros específicos da base de dados (Quebec e Mississippi)
quebec = base.loc[base['Type'] == 'Quebec']
mississippi = base.loc[base['Type'] == 'Mississippi']
```

```python
# Subgráfico (1 linha e 2 colunas) mostrando gráficos sobre cada região
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(x = quebec.conc, y = quebec.uptake).set_title("Quebec")
plt.subplot(1,2,2)
srn.scatterplot(x = mississippi.conc, y = mississippi.uptake).set_title("Mississippi")
plt.tight_layout()
```

```python
# Refrigerado e não refrigerado
ch = base.loc[base['Treatment'] == 'chilled']
nch = base.loc[base['Treatment'] == 'nonchilled']
```

```python
# Gráfico somente com 'chilled' e 'nonchilled'
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(x = ch.conc, y = ch.uptake).set_title("Chilled")
plt.subplot(1,2,2)
srn.scatterplot(x = nch.conc, y = nch.uptake).set_title("Non Chilled")
plt.tight_layout()
```
